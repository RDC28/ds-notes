```python
import pandas as pd

data = pd.read_csv('../EDA/export/data_after_EDA.csv')
```

```python
data.info() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

max_msrp_by_make = pd.DataFrame(data.groupby(by = ['Make'])['MSRP'].mean())
top_10_max_msrp_by_make = max_msrp_by_make.sort_values(by = 'MSRP', ascending = False)[0:10]

plt.plot(top_10_max_msrp_by_make / 100000)
plt.xlabel('Brand')
plt.ylabel('Price in Lakhs (dollars)')
plt.xticks(rotation = 90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

msrp_by_make_2017 = pd.DataFrame(data[data['Year'] == 2017].groupby(by = ['Make'])['MSRP'].mean())
top_10_msrp_by_make_2017 = msrp_by_make_2017.sort_values(by = 'MSRP', ascending = False)[0:10]

plt.bar(top_10_msrp_by_make_2017.index, height = top_10_msrp_by_make_2017['MSRP'] )
plt.xticks(rotation = 90)
plt.xlabel('Brand')
plt.ylabel('Average Sales in 2017 (dollars)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

top_10_brands = list(top_10_max_msrp_by_make.index)

top_10_brands_in_2009_2010_2015 = data.query("""`Make` in @top_10_brands and `Year` in (2009, 2010, 2015)""")
top_10_brands_in_2009_2010_2015 = pd.DataFrame(top_10_brands_in_2009_2010_2015.groupby(by = ['Make'])['MSRP'].mean())

plt.bar(x = top_10_brands_in_2009_2010_2015.index, height = top_10_brands_in_2009_2010_2015['MSRP'])
plt.xticks(rotation = 90)
plt.xlabel('Brands')
plt.ylabel('average MSRP in lakh dollars')
plt.title('Average MSRP of top 10 Brands in year 2009, 2010, 2015')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(top_10_max_msrp_by_make, color = 'black')
plt.bar(x = top_10_brands_in_2009_2010_2015.index, height = top_10_brands_in_2009_2010_2015['MSRP'], color = 'purple')
plt.xticks(rotation = 90)
plt.xlabel('Brands')
plt.ylabel('average MSRP in lakh dollars')
plt.title('Average MSRP of top 10 Brands in year 2009, 2010, 2015')
plt.show()
```

```python
plt.figure(figsize = [10, 8])
plt.pie(top_10_brands_in_2009_2010_2015['MSRP'],
        labels = top_10_brands_in_2009_2010_2015.index,
        explode = [0,0,0.25,0,0,0,0,0,0.25,0],
        autopct = '%.2f%%')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 2) # 2x2 grid

x = np.linspace(0, 10, 100)

ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title('Sine Wave')

ax[0, 1].plot(x, np.cos(x), color='red')
ax[0, 1].set_title('Cosine Wave')

ax[1, 0].scatter(x, x**2)
ax[1, 0].set_title('Parabola')

ax[1, 1].hist(np.random.randn(1000))
ax[1, 1].set_title('Histogram')

plt.tight_layout() # Adjust subplot parameters for a tight layout
plt.show()
```

max sales of a brand for a year

```python
import pandas as pd
import matplotlib.pyplot as plt

brand = iter(data['Make'].unique())
test = data.groupby(by = ['Make', 'Year'])['MSRP'].max()

fig, ax = plt.subplots(24, 2, figsize = (20, 100))

for row in range(24):
    for col in range(2):
        try:
            b = next(brand)
        except:
            break
        ax[row, col].bar(test[b].index, height = test[b])
        ax[row,col].set_title(b)
```
