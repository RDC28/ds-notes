```python
import pandas as pd

data = pd.read_csv('EDA/export/data_after_EDA.csv')
```

```python
data.info() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

max_msrp_by_make = pd.DataFrame(data.groupby(by = ['Make'])['MSRP'].mean())
top_10_max_msrp_by_make = max_msrp_by_make.sort_values(by = 'MSRP', ascending = False)[0:10]

plt.plot(top_10_max_msrp_by_make / 100000)
plt.xlabel('Brand')
plt.ylabel('Price in Lakhs (dollars)')
plt.xticks(rotation = 90)
plt.show()
```
