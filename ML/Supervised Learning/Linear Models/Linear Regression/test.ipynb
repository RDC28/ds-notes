```python
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data = np.array([1, 3, 4, 2, 5, 7])

data_mean = data.mean()
data_std= data.std()

print(f'mean: {data_mean}, std: {data_std}')
```

```python
data_scaled = (data - data_mean) / data_std
data_scaled.mean()
```
